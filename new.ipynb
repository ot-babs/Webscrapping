```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/phones/touch'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
product_name = []
for product in soup.find_all('a', {'class': 'title'}):
    product_name.append(product.get_text())
product_name
```

```python
ratings = []
for rat in soup.find_all('p', {'class': 'pull-right'}):
    ratings.append(rat.get_text())
ratings
```

```python
description = []
for desc in soup.find_all('p', {'class': 'description'}):
    description.append(desc.get_text())
description
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(product_name, pricelist, ratings, description)), columns=['Product Name', 'Price', 'Reviews','Description'])
df
```
