### Setup

````python
#WebScraping initialisaiton

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
url1 = 'https://fbref.com/en/squads/18bb7c10/Arsenal-Stats'
page1 = requests.get(url1)
soup1 = BeautifulSoup(page1.text, 'html.parser')
````

---
### Team Stats

````python
#Team name
teamdetail = soup1.find('div', {'data-template': 'Partials/Teams/Summary'}).text
splitdetail=teamdetail.split()
team = splitdetail[1]
print(team)

#Team Record
teamrecord = soup1.find('div', {'data-template': 'Partials/Teams/Summary'}).text
splitrecord=teamrecord.split()
record = splitrecord[8]
print(record)

# Governing Country
gcountry = soup1.find('div', {'data-template': 'Partials/Teams/Summary'}).text
splitcountry=gcountry.split()
country = splitcountry[68]
print(country)
````

---
### Player Stats

````python
#Name

playername = []
for player in soup1.find_all('th', {'scope': 'row'}):
    playername.append(player.get_text())
playername    

#Position
playerposition = []
for position in soup1.find_all('td', {'data-stat': 'position'}):
    playerposition.append(position.get_text())
playerposition

#Age
playerage = []
for age in soup1.find_all('td', {'data-stat': 'age'}):
    playerage.append(age.get_text())
playerage   
    
#Games Started
games_started = []
for starts in soup1.find_all('td', {'data-stat': 'games_starts'}):
    games_started.append(starts.get_text())
games_started   

#Minutes Played
playerminutes = []
for minutes in soup1.find_all('td', {'data-stat': 'minutes'}):
    playerminutes.append(minutes.get_text())
playerminutes


df1 = pd.DataFrame(list(zip(playername, playerposition, playerage, games_started, playerminutes)), columns=['Player','Position', 'Age', 'Games Started', 'Minutes Player'])
df1 = df1.iloc[:25]
df1
````
