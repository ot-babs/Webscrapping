### Completing The Lesson

```python
def attempts():
    
    targetbox = '/html/body/div[1]/div[3]/div[2]/div[3]/form/textarea'
    textbox = driver.find_element('xpath' ,targetbox)

    future = driver.find_element('xpath', '/html/body/div[1]/div[3]/div[2]/div[2]/span[2]').text
    splitfuture = future.split()
    futurelen = len(splitfuture)
    loop_count = futurelen + 2
    i=1
    while True:
        try:
            if i < loop_count: 
                first = driver.find_element('xpath', '/html/body/div[1]/div[3]/div[2]/div[2]/span[{}]'.format(i)).text
                textbox.send_keys(first)
                time.sleep(0.05)
                textbox.send_keys(" ")
                time.sleep(0.05)
                i +=1
            else:
                future = driver.find_element('xpath', '/html/body/div[1]/div[3]/div[2]/div[2]/span[2]').text
                splitfuture = future.split()
                futurelen = len(splitfuture)
                loop_count = futurelen + 2
                i = 1
        except:
            print('Game has now complete')
            WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.LINK_TEXT, "Click here"))).click()
            attempts()
            break
attempts()
```
