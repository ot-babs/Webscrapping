```python
#WebScraping initialisaiton

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
url1 = 'https://www.rightmove.co.uk/property-for-sale/find.html?locationIdentifier=REGION%5E87540&maxBedrooms=3&minBedrooms=3&propertyTypes=detached%2Csemi-detached%2Cterraced&includeSSTC=false&mustHave=&dontShow=&furnishTypes=&keywords='
page1 = requests.get(url1)
soup1 = BeautifulSoup(page1.text, 'html.parser')
```

```python
#Page1
propertyaddress = []
for property in soup1.find_all('div', {'class': 'propertyCard-details'}):
    propertyaddress.append(property.get_text())
propertyaddress

price1 = []
for price in soup1.find_all('div', {'class': 'propertyCard-priceValue'}):
    price1.append(price.get_text())
price1

agent1 = []
for agents in soup1.find_all('div', {'class': 'propertyCard-detailsFooter'}):
    agent1.append(agents.get_text())
agent1

df1 = pd.DataFrame(list(zip(propertyaddress, price1, agent1)), columns=['Property Address', 'Price', 'Property Card'])
df1


```
